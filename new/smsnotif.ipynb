```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
model = AutoModelForSequenceClassification.from_pretrained('/content/drive/MyDrive/bert-text-classification-model')
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
```

```python
department_mapping={
    0: "Cleanliness Department",
    1: "Service Department",
    2: "Defects Department"
}
```

```python
def route_complaint(complaint_text):
  inputs = tokenizer(complaint_text, return_tensors="pt", truncation=True, padding=True)
  with torch.no_grad():
    outputs = model(**inputs)
  predicted_category = torch.argmax(outputs.logits, dim=1).item()
  department = department_mapping[predicted_category]
  print(f"Complaint: '{complaint_text}' has been routed to the {department}.")

#have to be replaced with real-time address of complaints recieved from user

complaints = [
    "The train station was dirty and the bathrooms were not cleaned.",
    "The staff at the ticket counter was unhelpful and rude.",
    "The door on the train is broken and won’t close properly."
]

for complaint in complaints:
  route_complaint(complaint)
```
